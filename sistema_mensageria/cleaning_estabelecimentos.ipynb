## Load Data

```python
#Carregando ficheiros de enrtrada e saída
entrada = 'estabelecimentos_SP.csv'
saída = 'estabelecimentos_SP_limpos.csv.gz'
```

```python
#Lendo o arquivo de entrada
df = pd.read_csv(entrada, encoding = 'latin1')
```

**Colunas Categóricas e Numéricas**

```python
#Selecionando colunas catégoricas e numéricas
cat_cols = df.select_dtypes(include= ['object','category']).columns

num_cols = df.select_dtypes(include = np.number).columns
```

```python
print(f'Colunas Categóricas :{cat_cols}')
print(f'Colunas Numéricas : `{num_cols}')
```

```python
#Normalizando colunas categóricas
for col in cat_cols:
    if col in df.columns:  #preencher nulos com vazios, converter p str e maisculas e removendo acentos/espaços
        df[col] = df[col].fillna('').astype(str).apply(lambda x: unidecode(x).strip().upper()) 
```

```python
#Criar CNPJ como string (preservando zeros a esquerda)
df['CNPJ_BASICO'] = df['CNPJ_BASICO'].fillna('').astype(str).str.zfill(8) #colocando espaços nos nulos, convertendo p str e colocando 0 a esquerda.
df['CNPJ_ORDEM'] = df['CNPJ_ORDEM'].fillna('').astype(str).str.zfill(4)
df['CNPJ_DV'] = df['CNPJ_DV'].fillna('').astype(str).str.zfill(2)

```

```python
#Formar CNPJ completo com concatenação
df['CNPJ'] = df['CNPJ_BASICO'] + df['CNPJ_ORDEM'] + df['CNPJ_DV']
```

```python
#Remover CNPJS com menos de 14 dígitos (inválidos)
df = df[df['CNPJ'].str.len()==14]
```

```python
#Remover duplicados com base apenas no CNPJ
df = df.drop_duplicates(subset='CNPJ')
```

```python
#Converter colunas de datas
col_datas = ['DATA_SITUACAO_CADASTRAL', 'DATA_INICIO_ATIVIDADE', 'DATA_SITUACAO_ESPECIAL']

for col in col_datas:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce', format='%Y%m%d')
```

```python
#Salvar ficheiro limpo
df.to_csv(saída, index=False, encoding='utf-8',  compression='gzip')
```

```python
#Salvar em parquet para otimizar o funcionamento
df.to_parquet('estabelecimentos_SP_limpos.parquet', 
              index=False, 
              engine='pyarrow',  # especificar o engine
              compression='gzip')  # compactar também o Parquet
```
